```python
import os
import pandas as pd
```

```python
# Base folder
folder_path = './data'  # Use './data' if you're running notebook from project root
```

```python


# Create the list to collect all DataFrames
all_data = []

# Go through each year folder
for year in sorted(os.listdir('./data')):
    year_path = os.path.join('./data', year)
    if os.path.isdir(year_path):
        # Loop through each month folder (e.g., 'June2020TransBorderRawData')
        for month_folder in sorted(os.listdir(year_path)):
            month_path = os.path.join(year_path, month_folder)
            if os.path.isdir(month_path):
                # Loop through each CSV file in the month folder
                for file in os.listdir(month_path):
                    if file.endswith('.csv'):
                        file_path = os.path.join(month_path, file)
                        try:
                            # Read CSV with low_memory=False to handle mixed data types
                            df = pd.read_csv(file_path, low_memory=False)
                            df['year'] = year
                            df['month_folder'] = month_folder
                            df['source_file'] = file
                            all_data.append(df)
                        except Exception as e:
                            print(f"⚠️ Error loading {file_path}: {e}")

# After all files are loaded, combine into one DataFrame
if all_data:
    combined_df = pd.concat(all_data, ignore_index=True)
    print(f"✅ Loaded {len(all_data)} CSV files. Final shape: {combined_df.shape}")
    display(combined_df.head())
else:
    print("❌ No CSV files loaded. Please check folder structure and paths.")
```
